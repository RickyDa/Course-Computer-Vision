```python
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

### Zero Padding:
    padding the the given image.
    params:
        image - a Varibale that has the dimension of HxWxD
        H - Hieght
        W - Width
        D - Dimension
        k - Kernel dimesion
        stride - Stride of the kernel in convolution
    return:
        padded_image - padded image.

```python
def zero_padding(image, k, stride=1):
    w = image.shape[0]
    h = image.shape[1]
    p_w = int((w*(stride - 1) + k-stride)) 
    p_h = int((h*(stride - 1) + k-stride)) # stride = 1 => p = (k-1)/2 
    
    if len(image.shape) == 2: 
        padded_image = np.zeros([p_w+w,p_h+h])
        padded_image[1:-1,1:-1] = image
    elif len(image.shape) == 3:
        padded_image = np.zeros([p_w+w,p_h+h,image.shape[2]])
        padded_image[1:-1,1:-1,:] = image
    else:
        print('Expected to get 2D image')
        padded_image = None
   
    return padded_image    
```

```python
def load_image(path):
    return plt.imread(path)
```

```python
def plot_image(image):
    plt.figure(figsize=[20,10])
    plt.imshow(image,cmap='gray')
    plt.show()
```

```python
def convolve2d(image,kernel):
    k = kernel.shape[0] # Considering kernel is NxN (filter) matrix
    padded_image = zero_padding(image,k)
    output = np.zeros_like(image)
    if len(image.shape) < 3:
        d = 1
    else:
        d = image.shape[2]

    for z in range(d):
        for x in range(image.shape[0]):
            for y in range(image.shape[1]):
                output[x, y] = np.sum((kernel * padded_image[x: x+k, y: y+k]))
    return output
```

```python
image = load_image('./images/messi.jpg')
image = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
image = image/np.max(np.abs(image))
print(image.shape)
plot_image(image)
```

```python
new_image = convolve2d(image=image,kernel=np.array([[0, -1, 0], [-1, 8, -1], [0, -1, 0]])/4)
```

```python
plot_image(new_image)
```
