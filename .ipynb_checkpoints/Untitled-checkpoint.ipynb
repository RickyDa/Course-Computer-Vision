```python
from matplotlib import pyplot as plt
import numpy as np
import cv2
%matplotlib inline
```

### Zero Padding:
    padding the the given image.
#### params:
        image - a Varibale that has the dimension of HxWxD
        H - Hieght
        W - Width
        D - Dimension
        k - Kernel dimesion
        stride - Stride of the kernel in convolution
    
#### return:
        padded_image - padded image.

```python
def zero_padding(image, k, stride=1):
    w = image.shape[0]
    h = image.shape[1]
    p_w = int((w*(stride - 1) + k-stride)) 
    p_h = int((h*(stride - 1) + k-stride)) # stride = 1 => p = (k-1)/2 
    
    if len(image.shape) == 2: 
        padded_image = np.zeros([p_w+w,p_h+h],dtype='uint8')
        padded_image[1:-1,1:-1] = image
    elif len(image.shape) == 3:
        padded_image = np.zeros([p_w+w,p_h+h,image.shape[2]],dtype='uint8')
        padded_image[1:-1,1:-1,:] = image
    else:
        print('image dont have the right dims.')
        padded_image = None
        
    return padded_image
    
    
```

```python
def convolve2d(image,kernel):
    k = kernel.shape[0] # Considering kernel is NxN (filter) matrix
    padded_image = zero_padding(image,k)
    output = np.zeros_like(image)
    if len(image.shape) < 3:
        d = 1
    else:
        d = image.shape[2]
    for z in range(d):
        for x in range(image.shape[1]):
            for y in range(image.shape[0]):
                output[y, x] = (kernel * padded_image[y: y+k, x: x+k]).sum()
    return output
```

```python
def load_image(path):
    return plt.imread(path)
```

```python
def plot_image(image):
    plt.figure(figsize=[20,10])
    plt.imshow(image)
    plt.show()
```

```python
image = load_image('./images/messi.jpg')
plot_image(image)
ima_gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
new_image = convolve2d(image=image,kernel=np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]]))
```

```python
plot_image(new_image)
```
