```python
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

```python
image = plt.imread('./images/messi.jpg')
```

```python
cv2.__version__
```

```python
sobel = cv2.sqrt(cv2.addWeighted(cv2.pow(3, 2.0), 1.0, cv2.pow(3,2.0), 1.0, 0.0))
```

```python
ddepth = cv2.CV_64F
ksize = 3
scale= 1.0
```

```python
def norm(img):
    img = img.astype(float)
    img -= img.min()
    img /= img.max()
    return img
```

```python
gray = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)

dx = 0
dy = 1

sobelx = cv2.Sobel(norm(gray ),ddepth=ddepth,dx=dx,dy=dy,scale=scale,ksize=ksize)
plt.imshow(np.abs(sobelx),cmap='gray')
```
